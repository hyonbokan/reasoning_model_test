```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Context Scan', 'Ingestion Context Scan', 'ICS + CS']
exact_matches = [14, 16, 25]
partial_matches = [7, 13, 8]
total_matches = [exact_matches[i] + partial_matches[i] for i in range(len(labels))]

# Colors
exact_color = '#569cd6'
partial_color = '#ff9c29'

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
fig.patch.set_facecolor('#1e1e1e')  # Set background color for the figure
ax.set_facecolor('#1e1e1e')         # Set background color for the axes

# Create bar chart
bars1 = ax.bar(labels, exact_matches, color=exact_color, label='Exact Matches')
bars2 = ax.bar(labels, partial_matches, bottom=exact_matches, color=partial_color, label='Partial Matches')

# Annotate totals with bold and larger font
for i, total in enumerate(total_matches):
    ax.text(i, total + 1, str(total), ha='center', va='bottom', color='white', fontsize=14, fontweight='bold')

# Grid and labels
ax.set_ylabel('Number of Findings', color='white', fontsize=14)
ax.set_title('Number of Findings Comparison', color='white', fontsize=16)
ax.tick_params(colors='white', labelsize=14)
ax.legend(facecolor='#1e1e1e', edgecolor='white', labelcolor='white', fontsize=12)

# Add grid behind the bars
ax.yaxis.grid(True, color='gray', linestyle='--', linewidth=0.5)
ax.set_axisbelow(True)

# Increase Y-axis limit for spacing above bars
ax.set_ylim(0, max(total_matches) + 10)

plt.tight_layout()
plt.show()
```
